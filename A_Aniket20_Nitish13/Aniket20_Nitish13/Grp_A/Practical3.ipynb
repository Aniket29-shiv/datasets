#Basic Statistics - Measures of Central Tendencies and Variance


Write a Python program to display some basic statistical details like percentile, mean,
standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris-versicolor’
of iris.csv dataset.

```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Aniket29-shiv/datasets/main/Iris.csv")
```

```python
print('Iris-setosa')
setosa = data['Species'] == 'Iris-setosa'
print(data[setosa].describe())
```

```python
print('\nIris-versicolor')
setosa = data['Species'] == 'Iris-versicolor'
print(data[setosa].describe())
```

```python
print('\nIris-virginica')
setosa = data['Species'] == 'Iris-virginica'
print(data[setosa].describe())
```
